# Task 1(Creating Bar graph or histogram)

## Importing Necessary Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset 
df = pd.read_csv('E:\Free Lancing Projects\Prodigy_Infotech Internship tasks\Task_1\API_SP.POP.TOTL_DS2_en_csv_v2_31753.csv')
df.head()
```

## Preprocessing and creating Bar Graph

```python
# Striping any leading/trailing spaces from column names
df.columns = df.columns.str.strip()

# Selecting data for the year 2023
df_2023 = df[['Country Name', '2023']]

# Droping rows with missing data
df_2023 = df_2023.dropna()

# Create a bar chart for population distribution in 2020
plt.figure(figsize=(15, 8))
plt.bar(df_2023['Country Name'], df_2023['2023'])
plt.xlabel('Country')
plt.ylabel('Population')
plt.title('Population Distribution by Country in 2023')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

## Bar Graph for top 20 Country Name during year 2023

```python
'''
Since the above graph is too congested let us create an easy to understand bar graph of top 20 country name with highest population during the year 2023
'''

# Sort by population and take the top 20 countries
df_2023 = df_2023.sort_values(by='2023', ascending=False).head(20)

# Increase figure size and use horizontal bar chart
plt.figure(figsize=(10, 8))  # Increase the height for better readability
plt.barh(df_2023['Country Name'], df_2023['2023'], color='skyblue')
plt.xlabel('Population')
plt.ylabel('Country')
plt.title('Top 20 Countries by Population in 2023')

# Invert the y-axis to show the largest population at the top
plt.gca().invert_yaxis()

plt.tight_layout()  # Adjust layout to fit everything
plt.show()
```

## Creating Histogram

```python
# Create a histogram for population values in 2023
plt.figure(figsize=(8, 6))
plt.hist(df_2023['2023'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Population Size')
plt.ylabel('Frequency')
plt.title('Population Distribution in 2023')
plt.show()
```

## Code which provides line chart based on country name

```python
# List of all years
years = df.columns[4:]

# Function to plot population trend for a given country
def plot_country_population(country_name):
    # Check if the country exists in the dataset
    if country_name in df['Country Name'].values:
        # Filter the data for the given country
        country_data = df[df['Country Name'] == country_name]
        
        # Extract population data for all years
        population_values = country_data[years].values.flatten()

        # Create the line chart
        plt.figure(figsize=(15, 8))
        plt.plot(years, population_values, marker='o', color='b', linestyle='-', label=country_name)
        plt.xlabel('Year')
        plt.ylabel('Population')
        plt.title(f'Population Trend for {country_name}')
        plt.xticks(rotation=90)
        plt.grid(True)
        plt.tight_layout()
        plt.legend()
        plt.show()
    else:
        print(f"No data available for {country_name}")

# Loop to continuously ask for country names
while True:
    country_name = input("Enter the country name (or type 'exit' to quit): ")
    if country_name.lower() == 'exit':
        print("Exiting the program.")
        break  # Exit the loop if user types 'exit'
    
    # Call the function to plot the population data
    plot_country_population(country_name)
```
